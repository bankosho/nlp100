## 60. KVSの構築

Key-Value-Store (KVS) を用い，アーティスト名（name）から活動場所（area）を検索するためのデータベースを構築せよ

---
leveldb：
+ 一种key-value的数据库
+ key和value必须是byte列 利用encode()和decode()进行str和byte列的转换 
+ a.put(key,value)/Get(key)/Delete(key)  

---

```python
import json
import plyvel
from IPython.display import Image, display_png
```

```python
db=plyvel.DB('name_area', create_if_missing=True)
```

```python
with open("/Users/banootorishou/artist.json") as f: 
    for line in f:
        dic=json.loads(line) 
        name=dic.get("name")+'('+str(dic.get("id"))+')'
        area=dic.get("area",'') 
        db.put(name.encode(),area.encode())
```

```python
len(list(db))
```

## 61. KVSの検索

60で構築したデータベースを用い，特定の（指定された）アーティストの活動場所を取得せよ．

```python
artist=input("the name of artist:")
for key,value in db.iterator(prefix=(artist+'(').encode()):
    print("{}の活動場所は{}".format(key.decode(),value.decode()))
```

## 62. KVS内の反復処理

60で構築したデータベースを用い，活動場所が「Japan」となっているアーティスト数を求めよ．

```python
sum([1 for key,value in db if value.decode()=='Japan'])
```

## 63. オブジェクトを値に格納したKVS

KVSを用い，アーティスト名（name）からタグと被タグ数（タグ付けされた回数）のリストを検索するためのデータベースを構築せよ．さらに，ここで構築したデータベースを用い，アーティスト名からタグと被タグ数を検索せよ．

```python
db1=plyvel.DB('name_tags', create_if_missing=True)
```

```python
with open("/Users/banootorishou/artist.json") as f:
    for line in f:
        dic=json.loads(line)
        key=dic.get("name")+'('+str(dic.get("id"))+')'
        value=dic.get("tags",[])
        db1.put(key.encode(),json.dumps(value).encode()) #把value（list）变成json形式的然后再编码
```

```python
def retrieval():
    name=input("The name of artist")
    for key,value in db1.iterator(prefix=(name+'(').encode()):
        print("{}のタグ情報は:".format(key.decode()))
        if len(json.loads(value.decode()))>0:
            for x in json.loads(value.decode()):
                print("  {}({})".format(x.get("value"),x.get("count")))
        else:
            print("  タグ情報ありません")
```

```python
retrieval()
```

## 64. MongoDBの構築

アーティスト情報（artist.json.gz）をデータベースに登録せよ．さらに，次のフィールドでインデックスを作成せよ: name, aliases.name, tags.value, rating.value

```python
import pymongo
```

```python
client = pymongo.MongoClient()
db = client.test_database
artist = db.artist
```

```python
artist
```

```python
with open("/Users/banootorishou/artist.json") as f:
    bulk=[]
    for line in f:
        bulk.append(json.loads(line))
    artist.insert_many(bulk)
```

```python
artist.estimated_document_count()
```

```python
artist.create_index([('name', pymongo.ASCENDING)])
artist.create_index([('liases.name', pymongo.ASCENDING)])
artist.create_index([('tags.value', pymongo.ASCENDING)])
artist.create_index([('rating.value', pymongo.ASCENDING)])
```

```python
sorted(list(artist.index_information()))
```

## 68. ソート

"dance"というタグを付与されたアーティストの中でレーティングの投票数が多いアーティスト・トップ10を求めよ

```python
#col.find().sort('pokopoko', pymongo.DESCENDING)
```

```python
i=0
for doc in artist.find({'tags.value': 'dance'}).sort('rating.count',pymongo.DESCENDING):
    print("{}({})".format(doc["name"],doc["id"]))
    i+=1
    if i==10:break
```
